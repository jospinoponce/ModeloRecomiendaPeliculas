## EDA (Exploratory Data Analysis)

```python
# Se carga los dos datasets creados

eda_0=pd.read_csv("../Datasets/df.csv")
eda_1=pd.read_parquet("../Datasets/df_1.parquet")
```

```python
eda_0.head(3)
```

El dataset eda_1 tiene las puntuaciones que usuarios han generado a las peliculas. <br>
Se toma el promedio de las calificaciones de las peliculas agrupadas por por su ID.

```python
eda_1.head(11)
```

Se concatena los dos datasets eda_0 y eda_1 para formar un dataframe unico df_EDA. 

```python
# Se toma las variables a concatenar de eda_0

eda_0=eda_0[["id", "type", "release_year", "listed_in", "title"]]
eda_0.head(3)
```

```python
# Se realiza el join de los dos Datasets por la variable 'id'

df_EDA=pd.merge(eda_1,eda_0, on='id')
```
