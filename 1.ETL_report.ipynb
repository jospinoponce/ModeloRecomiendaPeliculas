## 1. Peliculas y series presentes en las plataformas
### ETL PROCESAMIENTO  CSV 'PLATAFORMAS'

```python
# Se importan los csv y se generan cuatro (4) dataframes

amazon=pd.read_csv("../Datasets/amazon_prime_titles.csv")
netflix=pd.read_csv("../Datasets/netflix_titles.csv")
disney=pd.read_csv("../Datasets/disney_plus_titles.csv")
hulu=pd.read_csv("../Datasets/hulu_titles.csv")
```

Conforme a los requerimientos del proyecto. Se añade la variable 'id'. Esta representa valor unico de cada registro.<br>
Tambien se genera la variable plataforma.  

```python
# Variable 'id'

amazon.insert(0, 'id', ('a'+amazon['show_id']))
netflix.insert(0, 'id', ('n'+netflix['show_id']))
disney.insert(0, 'id', ('d'+disney['show_id']))
hulu.insert(0, 'id', ('h'+hulu['show_id']))

# Variable 'plataforma'

amazon['plataforma']='amazon'
netflix['plataforma']='netflix'
disney['plataforma']='disney'
hulu['plataforma']='hulu'
```

Se crea un solo dataframe realizando una concatenación de los cuatro previamente creados. <br>
Se da formato al dataframe.

```python
df = pd.concat([amazon, netflix, disney, hulu], axis=0)
```

```python
f"dataframe df (rows,columns):{df.shape}"
```

```python
# Se resetea los indices

df.reset_index(drop=True, inplace=True)
```

```python
df.sample(10)
```

```python
# Se valida duplicados

f"la cantidad de valores duplicados es: {df.duplicated().sum()}"
```

```python
# tipo de formato de las variables

df.dtypes
```

```python
# Se divide la variable 'duration' en 'duration_int' y 'duration_type'

df['duration']=df['duration'].astype(str)

df['duration_int']= df['duration'].apply(lambda x: x.split()[0])

df['duration_type'] = df['duration'].apply(lambda x: "season" if 'Season'in x or 'Seasons' in x else "min")

df=df.drop(['duration'], axis=1)
```

```python
# Se modifica la variable 'date_added' de modo que se observen las fechas como AAAA-mm-dd

df['date_added']=pd.to_datetime(df['date_added'])
df['date_added'] = df['date_added'].dt.strftime("%Y-%m-%d")
```

```python
df.sample(29)
```

```python
# Validacion de datos nulos

df.isnull().sum().sort_values(ascending=False)
```

```python
# los nulos de la variable 'rating' se rellenan con la letra R

df[['rating']]= df[['rating']].fillna('R')

# Los nulos del dataframe se rellenan con la palabra "not_data"

df[['country', 'date_added','director','cast','description']]=df[['country', 'date_added',
'director','cast','description']].fillna('not_data')
```

```python
# Se muestra todos los campos de texto del dataframe en minisculas. (requerimiento del proyecto)

df=df.applymap(lambda x: x.lower() if type(x) == str else x)
```

```python
# Se observan datos que van acorde a la variable 'rating' 

df['rating'].value_counts()
```

```python
# Se imputa con "not_data" los campos de la variable 'rating' que presentan datos como: min, season, seasons

df['rating'] = df['rating'].apply(lambda x: "not_data" if 'min'in x or 'season' in x or 'seasons' in x else x)
```

```python
df.sample(9)
```

```python
# La variable 'listed_in' cuenta con muchos valores catégoricos. Se agrupa y se generan solamente 9. 

df['listed_in'].value_counts().head(30)
```

```python
# Se generan las nuevas categorias agrupando por palabras clave

# Comedy

df['listed_in'] = df['listed_in'].apply(lambda x: "comedy" if 'comedie' in x 
else ("comedy" if 'comedy' in x else x))

#horror y suspence

df['listed_in'] = df['listed_in'].apply(lambda x: "horror & suspence"  if 'horror' in x 
else ("horror & suspence" if 'suspense' in x else x))

#action

df['listed_in'] = df['listed_in'].apply(lambda x: "action" if 'action' in x 
else ("action" if 'adventure'in x else ("action" if 'crime'in x else ("action" if 'war' in x else x) )))
```

```python
# tv shows

df['listed_in'] = df['listed_in'].apply(lambda x: "family" if 'family' in x 
else ("family" if 'latino' in x else ("family" if 'serie' in x else ("family" if 'movies' in x else
("family" if 'unscripted' in x else ("family" if 'variety' in x else ("family" if 'young' in x else
("family" if 'lgbtq' in x else("family" if 'faith' in x else x)))))))))

df['listed_in'] = df['listed_in'].apply(lambda x: "tv shows" if 'tv shows' in x 
else ("tv shows" if 'sports' in x else ("tv shows" if 'music' in x else ("tv shows" if 'art'
in x else ("tv shows" if 'reality'in x else ("tv shows" if 'thriller' in x else ("tv shows" if 'food' in x else 
"tv shows" if 'show' in x else ("tv shows" if 'news' in x else ("tv shows" if 'lifestyle' in x else 
("tv shows" if 'fitness' in x else ("tv shows" if 'special'in x else ("tv shows" if 'international' in x else 
("tv shows" if 'family' in x else x)))))))))))))
```

```python
# documental

df['listed_in'] = df['listed_in'].apply(lambda x: "documentary" if 'documentarie' in x 
else ("documentary" if 'documentary' in x else ("documentary" if 'docu' in x else 
("documentary" if 'nature'in x else ("documentary"if 'histo' in x else x)))))

# kids
df['listed_in'] = df['listed_in'].apply(lambda x:"animation, kids" if 'animation'in x else ("animation, kids"if
"kids' tv" in x else ("animation, kids" if 'kids' in x else ("animation, kids" if 'anime' in x
else ("animation, kids" if 'children' in x else x)))))

# Romance

df['listed_in'] = df['listed_in'].apply(lambda x: "romance" if 'romance' in x 
else ("romance" if 'romantic' in x else x))

# drama
df['listed_in'] = df['listed_in'].apply(lambda x: "drama" if 'dramas' in x 
else ("drama" if 'drama' in x else ("drama" if 'western' in x else x)))

# science fiction

df['listed_in'] = df['listed_in'].apply(lambda x: "science fiction" if 'fantasy' in x 
else ("science fiction" if 'sci-fi' in x else ("science fiction" if 'science' in x else x)))
```
